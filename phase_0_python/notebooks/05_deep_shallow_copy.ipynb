# Show deep vs shallow copy using lists/dicts and prove difference via mutation.


Shallow copy → copies the container, shares the inner objects

Deep copy → copies the container and the inner objects

```python
# INPUTS:
# - nested list or dict

# OUTPUTS:
# - printed evidence of mutation difference

# INVARIANTS:
# - original object identity matters
# - mutation must target inner object

# SIDE EFFECTS:
# - intentional mutation for demonstration
```

```python
dict = {
    "parth" : [1,2,3,4],
    "tiwari" : [5,6,7,8],
    "1": [9,10,11,12]
}
dict
```

```python
list = [[1,2,3],[4,5,6],[7,8,9]]
list
```

```python
import copy
list2 = copy.deepcopy(list)
list2[0][2] = 99
list2, list
```

```python
list3 = copy.copy(list)
list3[0][2] = 123
list3,list
```
