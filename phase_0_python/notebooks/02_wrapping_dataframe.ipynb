“Wrap a pandas DataFrame in a class so that:

it stores the DataFrame

len(dataset) gives number of rows

dataset[i] returns the i-th row

printing the object shows a short summary”

the core object is a class

```python
# INPUTS: pandas DataFrame

# OUTPUTS: Dataset object that wraps the DataFrame

# INVARIANTS: - The underlying DataFrame is not modified
#          - Dataset length always equals number of rows

# SIDE EFFECTS:- None
```

```python
import pandas as pd
class Dataset:
    def __init__(self, df):
        self.df = df

    def __len__(self):
        return self.df.shape[0]

    def __getitem__(self, idx):
        return self.df.iloc[idx]

    def __repr__(self):
        rows, cols = self.df.shape
        return f"Dataset(rows={rows}, cols={cols})"
    
df = pd.read_csv('2.csv')

dataset = Dataset(df)
print(dataset.df.head())
print(len(dataset))
print(dataset[2])
print(dataset)
```
