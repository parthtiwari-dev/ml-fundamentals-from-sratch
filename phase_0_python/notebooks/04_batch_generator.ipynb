```python
# INPUTS:
# - X (array-like or DataFrame)
# - y (array-like or Series)
# - batch_size (int)

# OUTPUT:
# - generator yielding (X_batch, y_batch)

# INVARIANTS:
# - X and y alignment preserved
# - No sample dropped
# - Original data not modified

# SIDE EFFECTS:
# - None
```

```python
# - get number of samples
# - create list of indices
# - shuffle indices
# - for start in range(0, N, batch_size):
#     - batch_indices = indices[start : start + batch_size]
#     - X_batch = X.iloc[batch_indices]
#     - y_batch = y.iloc[batch_indices]
#     - yield X_batch, y_batch
```

```python
import pandas as pd
import numpy as np
import random

def batch_generator (X, y, batch_size):
    N = len(X)
    idx = np.arange(N)
    np.random.shuffle(idx)    
    for s_i in range(0, N, batch_size):
        e_i = s_i + batch_size
        batch_idx = idx[s_i:e_i]
        X_batch = X.iloc[batch_idx]
        y_batch = y.iloc[batch_idx]
        yield X_batch, y_batch
```
