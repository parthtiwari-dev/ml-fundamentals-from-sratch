#  Implement your own train_test_split using pure Python indices.

split dataset into 80:20 ratio

must use the logic of splitting randomly through out the data like how sklearn library would do 

```python
# INPUTS: pandas DataFrame

# OUTPUTS: 2 pandas DataFrame

# INVARIANTS: - The DataFrame is not modified
#          - the 2 Dataset combined length always equals to rows of original dataset

# SIDE EFFECTS:- None
```

```python
"""
N = number of rows
indices = [0, 1, 2, ..., N-1]
shuffle(indices)
split_point = int(0.8 * N)
train_indices = indices[:split_point]
test_indices = indices[split_point:]
"""
```

```python
import pandas as pd 
import random
random.seed(42)
def train_test_split(df):
    N = df.shape[0]
    idx = random.sample(range(0, N), N)
    split =  int(0.8 * N)
    train_indices = idx[:split]
    test_indices = idx[split:]
    df_train = df.iloc[train_indices]
    df_test = df.iloc[test_indices]

    return df_train,df_test

df = pd.read_csv('2.csv')

df_train, df_test = train_test_split(df)
```
